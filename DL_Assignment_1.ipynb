```
import pandas as pd
import numpy as np
```

```
df=pd.read_csv("/content/auto.csv",names=["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
 "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
 "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
 "peak-rpm","city-mpg","highway-mpg","price"])
```

```
df.head()
```

```
df.replace(to_replace="?",value=np.nan,inplace=True)
```

```
df.isnull().sum()
```

```
df.dropna(subset=['price'],inplace=True)
```

```
df.isnull().sum()
```

```
df['normalized-losses'].fillna("99999",inplace=True)
df['normalized-losses']=df['normalized-losses'].astype(np.int)
df['normalized-losses'].replace(to_replace=99999,value=np.nan,inplace=True)
df['normalized-losses'].fillna(df['normalized-losses'].mean(),inplace=True)
```

```
df['bore'].fillna("99999",inplace=True)
df['bore']=df['bore'].astype(np.float)
df['bore'].replace(to_replace=99999,value=np.nan,inplace=True)
df['bore'].fillna(df['bore'].mean(),inplace=True)
```

```
df['stroke'].fillna("99999",inplace=True)
df['stroke']=df['stroke'].astype(np.float)
df['stroke'].replace(to_replace=99999,value=np.nan,inplace=True)
df['stroke'].fillna(df['stroke'].mean(),inplace=True)
```

```
df['horsepower'].fillna("99999",inplace=True)
df['horsepower']=df['horsepower'].astype(np.float)
df['horsepower'].replace(to_replace=99999,value=np.nan,inplace=True)
df['horsepower'].fillna(df['horsepower'].mean(),inplace=True)
```

```
df['peak-rpm'].fillna("99999",inplace=True)
df['peak-rpm']=df['peak-rpm'].astype(np.float)
df['peak-rpm'].replace(to_replace=99999,value=np.nan,inplace=True)
df['peak-rpm'].fillna(df['peak-rpm'].mean(),inplace=True)
```

```
df['num-of-doors'].fillna(df['num-of-doors'].mode()[0],inplace=True)
```

```
df.reset_index
```

```
df.head()
```

```
df.dtypes
```

```
df['price']=df['price'].astype(np.float)
```

```
df.dtypes
```

```
df['city-L/100km']=235/df['city-mpg']
```

```
df['highway-L/100km']=235/df['highway-mpg']
```

```
df.head()
```

```
length_max=np.max(df['length'])
height_max=np.max(df['height'])
width_max=np.max(df['width'])

df['length']=df['length']/length_max
df['height']=df['height']/height_max
df['width']=df['width']/width_max
```

```
import matplotlib.pyplot as plt
plt.hist(df['horsepower'])
```

```
df.loc[df['horsepower']<100,'horsepower-binned']="Low"
df.loc[(df['horsepower']>100) & (df['horsepower']<=200),'horsepower-binned']="Medium"
df.loc[df['horsepower']>200,'horsepower-binned']="High"
```
